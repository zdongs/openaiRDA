````python

import json
import re

def replace_json(text):
    block = re.match(
        r"{.*?}", text, re.DOTALL
    )  # re.DOTALL 使 . 匹配任何字符
    if block is None:
        return {}
    # 将块转换为JSON格式
    json_ = json.loads(block[0])
    return json_


nollm = []
for i in data["岗位要求"].drop_duplicates(keep="first").tolist(): # keep="first" 保留第一次出现的值 
    if i in cache_dic:
        continue
    else:
        nollm.append(i)

len(nollm)
````
